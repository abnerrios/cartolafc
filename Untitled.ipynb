```python
from sklearn.model_selection import train_test_split

feature_names = ['field_area','venue','goals','assists','tackles_won','shots_total','shots_on_target','fouled','own_goals','cards_yellow','fouls',
'opponent_clean_sheets','opponent_possession','opponent_goals','opponent_goals_per_shot','opponent_tackles_won',
'opponent_fouled','opponent_shots_total','opponent_shots_on_target','opponent_cards_red','opponent_cards_yellow','opponent_fouls',
'opponent_saves','opponent_goals_against','opponent_goals_per_shot_against',
'opponent_shots_on_target_against','opponent_shots_total_against']
# inputs
X = data[feature_names]
# field to predict
y = data[['score']]
```

```python
X = pd.get_dummies(X, prefix='',drop_first=True)
X.fillna(0.0,inplace=True)
```

```python
# split da base
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.7, test_size=.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression

rfr = RandomForestRegressor(n_estimators=75,
                            bootstrap = True,
                            random_state=42)
rfr.fit(X_train, y_train)
```

```python
print('R^2 Training Score: {:.2f} \nR^2 Test Score: {:.2f}'.format(
                                                rfr.score(X_train, y_train),
                                                rfr.score(X_test, y_test))) 
```

```python
from sklearn.metrics import r2_score
from rfpimp import permutation_importances

def r2(rf, X_train, y_train):
    return r2_score(y_train, rf.predict(X_train))

perm_imp_rfpimp = permutation_importances(rfr, X_train, y_train, r2)
```
